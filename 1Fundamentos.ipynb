SLICING DE SECUENCIAS (PARTES DE UN ESTRUCTURA DE DATOS)
* Seleccion de un conjunto de elementos ordenados que pertenecen a una secuencia
* Se realiza mediante la notacion secuencial [inicio:termino:orden]
  * Inicio: indica el termino de donde se parte la extraccion (base 0). SI SE OMITE SE TOMA EL 0 COMO PARAMETRO
  * Termino: indica el termino de donde se finaliza la extraccion, la extraccion se realiza siempre menos un termino (base -1). SI SE OMITE EXTRAE HASTA EL ULTIMO TERMINO
  * Orden: tamaño del salto en que se realiza la extraccion. si se omite se toma el 1 como parametro

```python
lista1 = [1, 2, 3, 4, "a", "b"]
tupla1 = (1, 2, 3, 4, "a", "b")
tupla_nombres = ("uno", "dos", "tres", "cuatro", "letra a", "letra b")
str1 = "los strings son una secuencia de carateres"

print(tupla[0:4:1])
print(lista[0:4:2])
print(str1[5:20])
print(tupla[:4:2])
print(lista[0:])
print(str1[1:15:2])

# LONGITUD DE SECUENCIAS    len
print(len(lista1))
print(len(tupla1))
print(len(str1))

#OBTENCION DEL INDICE EN QUE APARECE UN ELEMENTO POR PRIMERA VEZ      index
print(lista1.index(2))
print(tupla1.index(1))
print(str1.index("t"))

#OBTERNER EL NUMER DE VECES QUE SE REPITE UN ELEMENTO       count
print(lista1.count(2))
print(tupla1.count(1))
print(str1.count("t"))

#ASIGNACION DE VARIABLES A CADA ELEMENTO DE UNA SECUENCIAS
var1, var2, var3, var4, var5, var6 = tupla1
print(var1)
print(var2)
print(var3)
print(var4)
print(var5)
print(var6)

# UNION DE SECUENCIAS ORDENADAMENTE
tupla_lista = zip(tupla_nombres, lista1)
print(list(tupla_lista))
```

```python
dict_1 = {"clave uno": 1, "clave dos": 2}
print(dict_1)

dict_2 = {10:1, 2:2}
print(dict_2)

# CREACION DE DICCIONARIOS ANIDADOS
dict_anidado = {"clave_1": 1, "clave_2": {"clave_21":[5, 2, 8, 6], "clave_22": 5, }}
print(dict_anidado)

# CREACION DE DICCIONARIOS DESDE TUPLAS Y LISTAS (CLAVE- VALOR)
tupla_ = (1, 2, 3, 4, 5, 6)
lista_ = ["unos", "dos", "tres", "cuatro", "cinco", "seis"]
dic_union = dict(zip(lista_, tupla_))
print(dic_union)
```

### CONJUNTOS

* Unidimensional (de un unico nivel), ya no existen las claves
* Mutable
* De tamaño variable (al ser mutable es obvio)
* Desordenada
* Cuyos elementos son valores o secuencias
* Cuyos elementos son unicos (no hay elemento repertidos)

```python
set1 = {1, 2, 3, 4, 5, 6}
print(set1)

#CONVERSION DE UN CONJUNTO
lista_UNO =[1, 2, 3, 4, 5, 6]
setuno = set(lista_UNO)
print(type(setuno))

#UNION DE DOS CONJUNTOS
set1 = {1, 2, 3, 4, 5, 6}
set2 = {6, 2, 8, 11, 5, 9}
print(set1 | set2)

#INTERSECCION DE DOS CONJUNTOS
set1 = {1, 2, 3, 4, 5, 6}
set2 = {6, 2, 8, 11, 5, 9}
print(set1 & set2)

#DIFERENCIA DE DOS CONJUNTOS
set1 = {1, 2, 3, 4, 5, 6}
set2 = {6, 2, 8, 11, 5, 9}
print(set1 - set2)

#DIFERENECIA SIMETRICA DE DOS CONJUNTOS
set1 = {1, 2, 3, 4, 5, 6}
set2 = {6, 2, 8, 11, 5, 9}
print(set1 ^ set2)

# INSERCION O MODIFICACION DE UN ELEMENTO POR CLAVE
set1 = {1, 2, 3, 4, 5, 6}
set1.add(7) # nombre del conjunto, la funcion "add" y el valor que se quiere agregar
print(set1)

# COMPROBACION DE UN ELEMENTO DENTRO DEL DICCIONARIO
set1 = {1, 2, 3, 4, 5, 6}
1 in set1 # el valor que queremos comprobar, "in" y el nombre del conjunto

# COMPROBACION DE UN SUBCONJUNTO DEL DICCIONARIO
set1 = {1, 2, 3, 4, 5, 6}
set3 = {1, 2, 3}
set1 >= set3 # ponemos el nombre del conjunto mayor y depues el nombre del subconjunto

# COMPROBACION DE UN SUPERCONJUNTO DEL DICCIONARIO
set1 = {1, 2, 3, 4, 5, 6}
set3 = {1, 2, 3}
set1 <= set3 # ponemos el nombre del superconjunto mayor y depues el nombre del conjunto

# ELIMINACION DE UN ELEMENTO POR SU VALOR
set_DOS = {1, 2, 3, 4, 5, 6}
set_DOS.remove(5) # funcion "del",  el nombre del conjunto y su valor
print(set_DOS)

set_DOS = {1, 2, 3, 4, 5, 6}
set_DOS.discard(9) # cuando no estamos seguro de que exista utilizamos esta funcion, si no existe no pasa nada 
print(set_DOS)

# ELIMINACION DE TODOS ELEMENTOS DE UN CONUNTO
set1 = {1, 2, 3, 4, 5, 6}
set1.clear() # el nombre del conjunto y funcion "clear",  
print(set1)
```

### ESTRUCTURAS DE CONTROL

Las estructuras de control de flujo nos permiten la construccion de programas. Estas estructuras son identicos en funcionamiento al resto de los lenguajes de programacion.


CONDICIONALES (if, elif, else)
* Estas estructuras de control nos permiten elegir, en base a una condicion dada, la ejecucuion de diferentes bloques de codigo.

```python
precio_taxi = 10
precio_bus = 2.5
dinero_disponible = 1.5 
```

```python
# Esta forma de realizar la condicionante es muy reduntante y obliga a las condiciones bastantes complejas

if precio_taxi < dinero_disponible:
  print("Me voy en taxi")
if precio_bus < precio_bus and precio_taxi > dinero_disponible:
  print("Me voy en bus")
if precio_bus > dinero_disponible:
  print("Me voy andando")

# Una forma de reescribir es
if precio_taxi < dinero_disponible:
  print("Me voy en taxi")
elif precio_bus < dinero_disponible:
  print("Me voy en bus")
else:
  print("Me voy andando")
```

BUCLE - for
* Estas estructuras de control nos permiten llevar a cabo la ejecucion de un bloque de codigo para todos los elementos condenidos en una secuencia donde pueden estar involucrados o no el bloque de codigo.

```python
for i in range(0, 10):
  print("Hola Mundo")

for i in range(0, 10):
  print("Hola Mundo: " + str(i))

lista_b = ["uno", "dos", "tres", "cuatro", "cinco", "seis"]
for element in lista_b: # para cada elemento de mi lis x 
  print(element)         # imprimir el elemento

#A veces es muy comodo disponer al mismo tiempo del elemento sobre el que se itera y el indice que ocupa dentro del bucle. 
# Para ello podemos utiliza "in" funcion "enumarte"

lista_b = ["uno", "dos", "tres", "cuatro", "cinco", "seis"]
for index, element in enumerate(lista_b): # index es la poscicion en la que me encuentro y enumerate trabaja con los valores de la lista
  print("Elemento en indice " + str(index) + " : " + element) #imprime la posicion : mas el elemento en la lista correspondiente
```

BUCLE - while
* Estas estructuras de control nos permiten llevar a cabo la ejecucion de un bloque de codigo minestras se cumple una condicion dada.

```python
i = 0 #TIENE UNA VARIABLE
while i < 10: #UNA CONDICION 
  print(i)  # REALIZA LA FUNCION IMPRIMIR HASTA QUE SE MANTENGA LA CONDICION
  i += 1 # LLEVA UNA JERARQUIA
```

INTERRUPCION DE BUCLES (break y continue)
* Las sentencias break y continue permiten ahorrar el flujo normal de ejecuciion de un bucle. Mas concretamentes:
  * Break: Corta completamente el flujo de ejecucion del bucle y pasa el control a la setencia posterior del bucle
  * Continue: Corta la ejecucion de la iteracion actual y pasa el control a la siguiente sentencia del bucle

```python
lista_COPIA1 = ["uno", "dos", "tres", "cuatro", "cinco", "seis"]
for element in lista_COPIA1:
  if element == "cuatro":
    break  # Lo que hace es generar un corte en el codigo omitiendo los valores d que siguen despues de la condicion
  print(element) #dando por terminada la iteracion

lista_COPIA2 = ["uno", "dos", "tres", "cuatro", "cinco", "seis"]
for element in lista_COPIA2:
  if element == "cuatro":
    continue # Lo que hace es generar un salto en el codigo, omitiendo el valor que cumple con la condicion if
  print(element)
```

COMPRENSIONES DE LISTAS
* Tenemos a nuestra disposicion una estrucutra especial de control (semejeante a los bucles) que facilita mucho la escritura de codigo 

```python
lista_M = ["enero", "febrero", "marzo", "abril", "mayo", "junio"]
lista_V = []
for element in lista_M:
  if element[0] == "m": # si elmento incicia con m
    lista_V.append(element.upper()) #agregar a la lista_V pero con mayusculas
print(lista_V)

# PODEMOS CONSEGUIR EL MISMO RESULTA CON UN CODIGO MAS CONCISO Y DE FORMA MAS RAPIDO
lista_M = ["enero", "febrero", "marzo", "abril", "mayo", "junio"]
lista_V = [element.upper() for element in lista_M if element[0] == "m"] 
#carcateristica con la cual se va agregar "upper", el bucle "for" y la condicion que tiene que cumplir para añada "if"
print(lista_V)
```

  Los errores se generan en forma de excepciones (objetos en los que se incluye tanto el detalle del error y la fila de codigo. Es importante realizar una buena gestion de excepciones de forma que los errores esten siempre controlados de forma que los progrmaas sean robustos y claros:

```python
# ESTO ES IMPORTANTE POR QUE HAY VECES QUE QUEREMOS QUE LE CODIGO SIGA CORRIENTO, AUNQUE SE GENEREN ERRORES
# LOS CUALES PODEMOS IDENTIFICAR AL FINAL DE CADA BLOQUE
cadena_1 = "12.52cadena"
entero_1 = float(cadena_1)
```

```python
# SE PUEDEN CONTROLAR TODOS LOS ERRORES DE MANERA GENERICA
try: # dentro de la identacion se excribe el codigo que queremos correr
  cadena_1 = "cadena"
  entero_1 = float(cadena_1)
except: # dentro de este bloque se pone las medidas que se deben toma con respecto al error
  print("esto es un error")

# PUEDE SER MAS DETALLADO
try:
  cadena_1 = "cadena"
  entero_1 = float(cadena_1)
except Exception as e: # la funcion exception nos indica cual es el error
  print(e)
  print("esto es un error")
```

```python
# O SE PUEDE REALIZAR UN CONTROL DETALLADO PARA CADA ENTRADA
try:
  cadena_1 = "cadena"
  entero_1 = float(cadena_1)
except RuntimeError as e: # ERROR A MITAD DE CORRER EL CODIGO
  print(e)
  print("esto es un error")
except ValueError as e: #EL FORMATO DEL VALOR NO ES CORRECTO 
  print(e)
  print("esto es un error")
except Exception as e: # FORMATO GENERICO
  print(e)
  print("esto es un error")
```

### FUNCIONES
* Pueden entenderse como un nombre asignado a un bloque de codigo que permite su posterior invocacion 
* Dispone de un conjunto de parametros de entradas (opcinales) sobre los que trabajara el bloque de codigo interno.
* Ofrecen un resultado de de salida (opcinal) que podra ser utilizado por el codigo que invoca a ala funcion
* Permite encapsular codigo de forma que puede ser reutlizado en varios puntos de un programa.
* Evitan la necesidad de copiar y pegar el codigo.
* Facilita el mantenimiento y reduce los posible puntos de error.
* Permite "generalizar" codigo al trabajar sobre unos parametros de entrada que puden ser "variables".

```python
lista_W = ["enero", "febrero", "marzo", "abril", "mayo", "junio"]
lista_Z = []
for element in lista_W:
  if element[0] == "m": # si elmento incicia con m
    lista_Z.append(element.upper()) #agregar a la lista_V pero con mayusculas
print(lista_Z)

#SIN AMBARGO CUANDO QUEREMOS REPLICAR EL PROCEDIMIENTO MUCHAS ES CONVENIENTE CREAR UNA FUNCION

def fun_lista (lista, inicial, mayuscula): #def, nombre de la funcion y los parametros o varibles que vamos a ocupar
  lista_out = [] # lista que nos va devolver
  for element in lista: # nombre de la lista que va a analizar
    if element[0] == inicial:  # si inicia la con la letra que se indica "inicial" añade el elemento a la lista de salida
      if mayuscula:             # despues no regunta si lo queremos en may o min.
        lista_out.append(element.upper())
      else:
        lista_out.append(element.lower())
  return lista_out # nos devolvera una lista, la funcion return nos devuelve el resultado

print(fun_lista(lista_W, "m", True))

lista_P = ["mercurio", "venus", "tierra", "marte", "jupiter", "saturno", "neptuno"]
print(fun_lista(lista_P, "v", True))
print(fun_lista(lista_P, "m", True))
print(fun_lista(lista_P, "s", False))
```

ARGUMENTOS OPCIONALES
 
Podemos asignar valores por defecto a los parametros de una funcion

```python
def fun_lista (lista, inicial, mayuscula=True): # los argumentos son los que estan en el parentesis
  lista_out = []
  for element in lista:
    if element[0] == inicial:
      if mayuscula:
        lista_out.append(element.upper())
      else:
        lista_out.append(element.lower())
  return lista_out

# Al especificar el TRUE, por default se tranforman en Mayusculas, para pasar a minusculas hay que especificar

lista_P = ["mercurio", "venus", "tierra", "marte", "jupiter", "saturno", "neptuno"]
print(fun_lista(lista_P, "m"))
print(fun_lista(lista_P, "m", True))
print(fun_lista(lista_P, "m", False))
```

ESPECIFICAR EL NOMBRE DE LOS ARGUMENTOS EN LA LLAMADA

Se puede especificar el nombre de los argumentos de entrada en la llamada, eliminando la necesidad de mantener su orden(su funcionamiento por default

```python
#CUANDO SON MUCHOS ARGUMENTOS ES MAS FACIL ESPECIFICAR LOS ARGUMENTOS 
lista_P = ["mercurio", "venus", "tierra", "marte", "jupiter", "saturno", "neptuno"]
print(fun_lista(inicial="v", lista=lista_P)) # ESPECIFICA LOS ARGUMENTOS
```

RETORNO DE MULTIPLES ELEMENTOS

El retorno de las funciones sera unico (un unico valor que se podra asignar a una variable en el codigo que invoca) seria posible devolver una secuencia y recoger los resultados en multiples variables

```python
def fun_lista (lista, inicial, mayuscula=True): 
  lista_out = []
  for element in lista:
    if element[0] == inicial:
      if mayuscula:
        lista_out.append(element.upper())
      else:
        lista_out.append(element.lower())
  return (len(lista_out), lista_out)

lista_P = ["mercurio", "venus", "tierra", "marte", "jupiter", "saturno", "neptuno"]
num_elementos, elementos = fun_lista(lista_P, "m")
for i in range(0, num_elementos):
  print(elementos[i])
```
